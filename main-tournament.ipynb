```python
%load_ext autoreload
%autoreload 2

from hunt.tournament import Tournament
from hunt.game import DesignGame, Player
from hunt.player import RandomPlayer, MirrorPlayer
from hunt.game_theory import RiskDominancePlayer, ExpectedValuePlayer
from hunt.plotting import plot_panel, plot_pie_panel

all_players = [Player, RandomPlayer, MirrorPlayer, 
           RiskDominancePlayer, ExpectedValuePlayer]
```

```python
phase_1_solo = Tournament(
    players = all_players,
    games = [
        DesignGame([[[75, 88], [38, 100]]])
    ],
    num_reps=200
)
phase_1_solo.run()
```

```python
plot_panel(phase_1_solo, "random", "mirror")
plot_pie_panel(phase_1_solo, "random", "mirror")
```

```python
phase_1_full = Tournament(
    players = all_players,
    games = [
        DesignGame([[[75, 88], [63, 100]]]),
        DesignGame([[[75, 63], [38, 100]]]),
        DesignGame([[[50, 75], [25, 100]]]),
        DesignGame([[[50, 25], [-25, 100]]]),
        DesignGame([[[75, 88], [38, 100]]]),
        DesignGame([[[75, 63], [-38, 100]]]),
        DesignGame([[[50, 75], [-25, 100]]]),
        DesignGame([[[50, 25], [-175, 100]]]),
        DesignGame([[[75, 88], [71, 100]]]),
        DesignGame([[[75, 63], [63, 100]]]),
        DesignGame([[[50, 75], [42, 100]]]),
        DesignGame([[[50, 25], [25, 100]]])
    ],
    num_reps=200
)
phase_1_full.run()
```

```python
plot_panel(phase_1_full, "random", "mirror")
plot_pie_panel(phase_1_full, "random", "mirror")
```
